```python
# Function to extract top 3 cast members from the 'cast' column

#cast
def convert3(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            l.append(i['name'])
            counter+=1
        else:
            break
                
    return l
```

```python
# Apply the function to the 'cast' column

m['cast']=m['cast'].apply(convert3)
```

```python
# Function to extract director from the 'crew' column

def fetch_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
            break
    return l
```

```python
# def stem(text):
#     y=[]
#     for i in text.split():
#         y.append(ps.stem(i))
#     return " ".join(y)


def lemmatize(text):
    lemmatized_words = [lemmatizer.lemmatize(word) for word in text.split()]
    return " ".join(lemmatized_words)
```

```python
# new_df['tags']=new_df['tags'].apply(stem)

from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

# Initialize WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

# Define a function to lemmatize text
def lemmatize_text(text):
    lemmatized_words = [lemmatizer.lemmatize(word) for word in word_tokenize(text)]
    return " ".join(lemmatized_words)

# Apply lemmatization to the 'tags' column in new_df
new_df['tags'] = new_df['tags'].apply(lemmatize_text)
```

```python
# Import CountVectorizer to convert text into a matrix of token counts (Bag of Words)

from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()
```

```python
#forming vectors, using bag of words, using stemming
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
# Import cosine_similarity to calculate similarity between movies

from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Calculate similarity between movie vectors

similarity=cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:10]
```

```python
# Function to recommend similar movies based on cosine similarity

def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list =sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:10]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
import tkinter as tk
from fuzzywuzzy import process  # Ensure you have the fuzzywuzzy package installed

# Define a function that takes an input and generates some output
def recommend(movie):
    # Clean the input movie title by converting it to lowercase and stripping extra spaces
    movie = movie.lower().strip()
    
    # Use fuzzy matching to find the closest matching title from the dataset
    closest_match = process.extractOne(movie, new_df['title'].str.lower())
    
    if closest_match:
        movie_index = new_df[new_df['title'].str.lower() == closest_match[0]].index[0]
        distances = similarity[movie_index]
        movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:40]
        recommended_movies = []
        for i in movies_list:
            recommended_movies.append(new_df.iloc[i[0]].title)
        return recommended_movies
    else:
        return ["No matching movie found. Please try again."]

# Function to call the recommend function with the input value and display the output
def call_recommend(event=None):  # Add 'event=None' to handle key binding
    input_value = input_entry.get()
    recommended_movies = recommend(input_value)
    recommended_movies_text = "\n".join(recommended_movies)
    output_text.delete('1.0', tk.END)
    output_text.insert(tk.END, recommended_movies_text)

# Create the main window
root = tk.Tk()

# Set the title of the window
root.title("Movie Recommender")

# Set the background color of the window
root.configure(bg='#2d2d2d')

# Create a label to prompt the user to enter a movie name
input_label = tk.Label(root, text="Enter a movie name:", font=("Arial", 14), fg='#ffffff', bg='#2d2d2d')
input_label.pack()

# Create an entry widget for the user to enter a movie name
input_entry = tk.Entry(root, font=("Arial", 14), width=30)
input_entry.pack(pady=10)

# Bind the Enter key to the call_recommend function
root.bind('<Return>', call_recommend)  # Bind the Enter key to call_recommend

# Create a label to display the recommended movies
output_label = tk.Label(root, text="Recommended Movies:", font=("Arial", 14), fg='#ffffff', bg='#2d2d2d')
output_label.pack()

# Create a text box to display the recommended movies
output_text = tk.Text(root, font=("Arial", 14), height=10, width=50, fg='#ffffff', bg='#2d2d2d', bd=0)
output_text.pack()

# Create a button to call the recommend function with the input value and display the output
submit_button = tk.Button(root, text="Submit", font=("Arial", 14), fg='#ffffff', bg='#1db954', command=call_recommend)
submit_button.pack(pady=10)

# Center the window on the screen
root.eval('tk::PlaceWindow %s center' % root.winfo_toplevel())

# Start the main loop
root.mainloop()
```
